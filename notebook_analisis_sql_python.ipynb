El top de accidentes ocurridos en China son relacionados a incendios/fuego, hubo una locación donde fallecieron hasta 9 personas. El rubro donde ocurrieron mas accidentes fue en el sector construcción.

```python
import plotly.express as px
fig = px.bar(
    df,
    x ="Industry",
    title="Accidentes por sector")
fig.show()
```

El sector con mas accidentes es el de construcción triplicando aproximadamente la cantidad del segundo lugar el cual es minería de carbon. Muy cercano esta, en tercer lugar, la industria de hierro y acero.

```python
import plotly.express as px
fig = px.bar(
    df,
    x ="City",
    title="Accidentes por ciudad")
fig.show()
```

De todas las ciudades, las que presentan mas cantidad de accidentes de trabajo son las ciudades Tianjing y Bejing.

```python
fig = px.histogram(
    df,
    x="Number of Deaths",
    title="Distribución de personas fallecidas"
)

fig.show()
```

Hay 135 empresas que en el periodo de tiempo del registro solo han presentado accidentes de trabajo con lesiones. Por otro lado, 137 empresas que presentaron accidentes de trabajo con consecuencia de muertes, entre 1 a 9 muertos en cada evento ocurrido. Hay un tercer escenario, donde 7 empresas tuvieron accidentes de trabajo con impactó en cada evento, entre 10 a 29 trabajadores fallecidos.

```python
fig = px.histogram(
    df,
    x="Number of Injuries",
    title="Distribución de personas lesionadas"
)

fig.show()
```

```python
fig = px.box(
    df,
    y="Financial Penalty",
    title="Distribución de importe de la multa"
)
fig.show()
```

Por lo que se aprecia, las empresas con accidentes de trabajo o no han registrado su sanción económica por los accidentes de trabajo ocurridos o no ahn sido impuestas por el ente fisclaizador.

```python
duckdb.query("""
SELECT *
FROM df
LIMIT 10
""").to_df()
```

```python
duckdb.query("""
SELECT *
FROM df
WHERE Industry ='Construction'
""").to_df()
```

```python
duckdb.query("""
SELECT Industry,
COUNT(*) AS Total
FROM df
GROUP BY Industry
ORDER BY Total DESC
""").to_df()
```

```python
duckdb.query("""
SELECT "Company Ownership",
AVG("Financial Penalty") AS Promedio
FROM df
GROUP BY "Company Ownership"
ORDER BY Promedio DESC
""").to_df()
```
